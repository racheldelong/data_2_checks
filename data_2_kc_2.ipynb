```python
import pandas as pd
import re
```

```python
# import the 'horror movies' csv
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-22/horror_movies.csv")
```

```python
df.head(4)
```

```python
# remove rows without a listed budget
df = df.dropna(subset=['budget'])

# reset index
df = df.reset_index(drop=True)

df.tail(4)
```

```python
# convert release_date column to datetime
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True)

df.tail(4)
```
